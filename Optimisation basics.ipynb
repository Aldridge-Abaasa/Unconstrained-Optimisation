### Unconstrained Optimisation using Gradient Descent: 

$S = \begin{bmatrix}

### An expression can be derived in terms of and for the optimal step size for gradient descent applied to find the unconstrained minimum of a quadratic function to guide our investigation

$F(x) = \frac{1}{2}x^TSx$ 

### with a generic symmetric and positive-definite matrix .

### Note: to apply the gradient descent to find the unconstrained minimum of the quadratic function $F(x) = \frac{1}{2}x^TSx$, $x$ will have to be updated iteratively by taking steps in the negative direction of the gradient. The gradient of $F(x)$ can be calculated as follows:

$\nabla F(x) = Sx$

The optimal step size for gradient descent can be found by minimizing the function $F(x - \alpha\nabla F(x))$ with respect to $\alpha$. We can express this function as follows:

$F(x - \alpha\nabla F(x)) = \frac{1}{2}(x - \alpha Sx)^T S (x - \alpha Sx)$

Expanding the quadratic form and simplifying, we obtain:

$F(x - \alpha\nabla F(x)) = \frac{1}{2}x^TSx - \alpha x^TS^TSx + \frac{1}{2}\alpha^2x^TSSx$

To find the value of $\alpha$ that minimizes this function, we can take the derivative with respect to $\alpha$ and set it equal to zero:

$\frac{d}{d\alpha} [\frac{1}{2}x^TSx - \alpha x^TS^TSx + \frac{1}{2}\alpha^2x^TSSx] = -x^TS^TSx + \alpha x^TSS^TSx = 0$

Simplifying and solving for $\alpha$, we obtain:

$\alpha_{opt} = \frac{x^TSSx}{x^TSS^TSx}$

Substituting the matrix $S$, we have:

$\alpha_{opt} = \frac{x^T\begin{bmatrix}5 & -3 \ -3 & 5\end{bmatrix}x}{x^T\begin{bmatrix}25 & -15 \ -15 & 25\end{bmatrix}x}$

Simplifying, we get:

$\alpha_{opt} = \frac{5x_1^2 - 6x_1x_2 + 5x_2^2}{25x_1^2 - 30x_1x_2 + 25x_2^2}$

### Therefore, the optimal step size for gradient descent applied to find the unconstrained minimum of the quadratic function $F(x) = \frac{1}{2}x^TSx$ with matrix $S$ and initial point $x$ is given by the above expression.



### By considering the worst case scenario (worst $x_i$ ) for the decrease in the error $e_k = ||x_k - x^*||$from one step to the next, derive an expression for the optimal step size in that case in terms of the eigenvalues of S.


### To derive the optimal step size for the worst-case scenario for the decrease in the error $e_k = ||x_k - x^*||$, we can consider the case where $x_k$ is aligned with the eigenvector corresponding to the largest eigenvalue of $S$, which we denote as $\lambda_1$. In this case, we have $x_k = c_1 v_1$, where $v_1$ is the corresponding eigenvector and $c_1$ is a scaling factor.

### We can then consider the decrease in error from one step to the next as follows:

$e_k - e_k+1 = ||x_k - x^*|| - ||x_k+1 - x^*||$

### Substituting in the corresponding eigenvector and constant

$e_k - e_k+1 = ||c_1v_1 - x^*|| - ||c_1v_1 - \alpha S(c_1v_1) - x^*||$

### we can then represent S with its eigenvector 

$e_k - e_k+1 = ||c_1v_1 - x^*|| - ||(c_1 - \alpha \lambda_1c_1)v_1 - x^*||$

### To minimise $e_k+1$ with respect to $\alpha$, we focus on the expression $||(c_1 - \alpha \lambda_1c_1)v_1 - x^*||$ and aim to minimise it. Since $c_1$ and $\lambda_1$ are both positive, we can minimise the expression: 

$c_1 - \alpha \lambda_1 c_1 = 0$ which gives: 

$\alpha = 1 /\lambda_1$

### therefore: $e_k - e_k+1 = (1 - 1/\lambda_1) ||c_1v_1 - x^*||$

### Since we are considering the worst case scenario we aim to chose c_1 such that we can maximise $||c_1v_1 - x^*||$ therefore the optimal step size for this case will be $\alpha = 1 /\lambda_1$ where $\lambda_1$ is the largets eigenvalue of $S$.


### So What condition must $x_i$ satisfy to be in this worst-case scenario?

$x_i$ must correspond with the largets eigenvalue of S for the worst case scenario. 

### (However, it might be helpful to explicitly state that this step size guarantees convergence since 0 < $\frac{1}{λ₁} < $\frac{2}{λ₁})

### Hence Working with the specific matrix above and starting from compute six steps using gradient descent with the optimal step size.

Using the $S = \begin{bmatrix}
5 & -3 \\
-3 & 5 
\end{bmatrix}$ we can compute its eigenvalues. Eigenvalues are $\lambda_1 = 8$ and $\lambda_2 = 2$ and corresponding eigenvectors are $v_1 = [1, 1]$ and $v_2 = [-1, 1]$


With $x_0 = (3, 5)$ we compute the gradient of $F(x)$ at this initial point: 

$\nabla F(x) = Sx = S(3, 5) = (10, 4)$
We canuse the step size $\alpha = 1 / \lambda_1 = 1/8$ to update $x_0$: 

$x_1 = x_0 - \alpha \nabla F(x_0) = (3, 5) - 1/8(10, 4) = (2.125, 4.25)$

$x_2 = x_1 - \alpha \nabla F(x_1) = (2.125, 4.25) - 1/8 (8.5, 2) = (1.6094, 3.6914)$

$x_3 = x_2 - \alpha \nabla F(x_2) = (1.6094, 3.6914) - 1/8 (6.332, 0.531) = (1.2353, 3.2053)$

$x_4 = x_3 - \alpha \nabla F(x_3) = (1.2353, 3.2053) - 1/8 (4.684, -1.045) = (0.9738, 2.7294)$

$x_5 = x_4 - \alpha \nabla F(x_4) = (0.9738, 2.7294) - 1/8 (3.458, -1.946) = (0.8009, 2.3683)$

```python
import numpy as np

# Define the matrix S
S = np.array([[5, -3],
              [-3, 5]])

# Initial point
x0 = np.array([3, 5])

# Step size
alpha = 1/8

def gradient_descent_step(x, S, alpha):
    gradient = S @ x
    return x - alpha * gradient

# Perform 6 steps
x = x0
steps = [x.copy()]

for i in range(6):
    x = gradient_descent_step(x, S, alpha)
    steps.append(x.copy())
    print(f"x{i+1} = [{x[0]:.4f}, {x[1]:.4f}]")

# Calculate function values
def quadratic_function(x, S):
    return 0.5 * x.T @ S @ x

# Print function values
for i, x in enumerate(steps):
    f_value = quadratic_function(x, S)
    print(f"F(x{i}) = {f_value:.4f}")
```

##### Next we investigate the error ek in each step. By considering how the error decreases over 3 steps it is vital to verify that the iteration is converging linearly to the solution x*.

```python
import numpy as np

# Define matrix S
S = np.array([[5, -3], [-3, 5]])

# Define the true solution x* (which is [0,0] for this quadratic optimization)
x_star = np.array([0, 0])

# Initialize the starting point x0 and error list
x_k = np.array([3, 5])
errors = []

# Step size is 1/λ₁ = 1/8 (from part b)
alpha = 1/8

# Iterate 6 times
for k in range(6):
    # Compute error before update
    error = np.linalg.norm(x_k - x_star)
    errors.append(error)
    
    # Compute gradient
    gradient = S @ x_k
    
    # Update x_k using gradient descent
    x_k = x_k - alpha * gradient

# Add final error
error = np.linalg.norm(x_k - x_star)
errors.append(error)

# Print errors and compute ratios to verify linear convergence
print("Errors at each iteration:")
for i, e in enumerate(errors):
    print(f"e_{i} = {e:.6f}")

print("\nConvergence ratios (e_{k+1}/e_k):")
for i in range(len(errors)-1):
    ratio = errors[i+1]/errors[i]
    print(f"e_{i+1}/e_{i} = {ratio:.6f}")

# Plot the errors on a log scale to verify linear convergence
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.semilogy(range(len(errors)), errors, 'bo-', linewidth=2, markersize=8)
plt.grid(True)
plt.xlabel('Iteration k')
plt.ylabel('Error (log scale)')
plt.title('Error Convergence in Gradient Descent')
plt.show()
```

##### Working with the matrix above and starting from compute six steps using gradient descent with momentum and employing the values for the optimal step size and momentum parameter in terms of the eigenvalues:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define matrix S
S = np.array([[5, -3], [-3, 5]])

# Calculate eigenvalues of S
eigenvalues = np.linalg.eigvals(S)
lambda_1 = np.max(eigenvalues)  # = 8
lambda_2 = np.min(eigenvalues)  # = 2

# Calculate optimal parameters
alpha = 4 / ((lambda_1 + lambda_2))  # Optimal step size
beta = ((np.sqrt(lambda_1) - np.sqrt(lambda_2)) / (np.sqrt(lambda_1) + np.sqrt(lambda_2)))**2  # Momentum parameter

print(f"Optimal step size (α): {alpha:.6f}")
print(f"Momentum parameter (β): {beta:.6f}")

# Initialize variables
x_k = np.array([3.0, 5.0])  # Starting point
v_k = np.zeros(2)  # Initial velocity
iterations = []
x_history = [x_k.copy()]

# Perform 6 iterations of gradient descent with momentum
for k in range(6):
    # Compute gradient
    gradient = S @ x_k
    
    # Update velocity using momentum
    v_k = beta * v_k - alpha * gradient
    
    # Update position
    x_k = x_k + v_k
    
    # Store results
    x_history.append(x_k.copy())
    iterations.append({
        'iteration': k+1,
        'x': x_k,
        'gradient': gradient,
        'velocity': v_k
    })

# Print results
print("\nGradient Descent with Momentum iterations:")
print(f"x₀ = [{x_history[0][0]:.4f}, {x_history[0][1]:.4f}]")
for i, iter_data in enumerate(iterations):
    print(f"\nIteration {i+1}:")
    print(f"x_{i+1} = [{iter_data['x'][0]:.4f}, {iter_data['x'][1]:.4f}]")
    print(f"gradient = [{iter_data['gradient'][0]:.4f}, {iter_data['gradient'][1]:.4f}]")
    print(f"velocity = [{iter_data['velocity'][0]:.4f}, {iter_data['velocity'][1]:.4f}]")

# Visualize the optimization path
plt.figure(figsize=(10, 8))

# Create contour plot of the objective function
x1, x2 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = 0.5 * (5*x1**2 + 5*x2**2 - 6*x1*x2)

plt.contour(x1, x2, Z, levels=20, cmap='viridis')
plt.colorbar(label='F(x)')

# Plot optimization path
x_history = np.array(x_history)
plt.plot(x_history[:,0], x_history[:,1], 'ro-', label='Optimization Path', linewidth=2)
plt.plot(x_history[0,0], x_history[0,1], 'go', label='Starting Point', markersize=10)
plt.plot(0, 0, 'ko', label='Optimal Point', markersize=10)

# Plot eigenvectors
v1 = np.array([1, 1]) / np.sqrt(2)  # Normalized eigenvector for λ₁
v2 = np.array([-1, 1]) / np.sqrt(2)  # Normalized eigenvector for λ₂
plt.quiver(0, 0, v1[0]*4, v1[1]*4, angles='xy', scale_units='xy', scale=1, 
           color='r', label='Eigenvector λ₁=8', width=0.02)
plt.quiver(0, 0, v2[0]*4, v2[1]*4, angles='xy', scale_units='xy', scale=1,
           color='b', label='Eigenvector λ₂=2', width=0.02)

plt.grid(True)
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('Gradient Descent with Momentum\nQuadratic Function F(x) = ½xᵀSx')
plt.legend()
plt.axis('equal')
plt.show()
```

#### Regression using gradient descent: 

The optimisation problem: The optimization problem required to determine the parameters $a$, $b$, and $c$ that best fit the data using linear regression is to minimize the sum of squared errors between the predicted values and the actual values of the target variable $f_i$ over all the data points $i$. Specifically, the optimization problem is:

$min_{a,b,c} \sum_{i=1}^{n} [f_i - (a + bx_i + cy_i)]^2$

To solve the optimization problem using gradient descent without momentum, we need to compute the gradient of the objective function with respect to the parameters $a$, $b$, and $c$, and then update the parameters iteratively using the following update rule:

$a_{k+1} = a_{k} - \alpha \frac{\partial}{\partial a} J(a_k,b_k,c_k)$

$b_{k+1} = b_{k} - \alpha \frac{\partial}{\partial b} J(a_k,b_k,c_k)$

$c_{k+1} = c_{k} - \alpha \frac{\partial}{\partial c} J(a_k,b_k,c_k)$

where $J(a,b,c)$ is the objective function that we want to minimize, and $\alpha$ is the step size or learning rate.

For our specific problem, the objective function is:

$J(a,b,c) = \sum_{i=1}^{n} \left[ f_i - (a + b x_i + c y_i) \right]^2$

To compute the gradients, we have:

$\frac{\partial}{\partial a} J(a,b,c) = -2 \sum_{i=1}^{n} \left[ f_i - (a + b x_i + c y_i) \right]$

$\frac{\partial}{\partial b} J(a,b,c) = -2 \sum_{i=1}^{n} x_i \left[ f_i - (a + b x_i + c y_i) \right]$

$\frac{\partial}{\partial c} J(a,b,c) = -2 \sum_{i=1}^{n} y_i \left[ f_i - (a + b x_i + c y_i) \right]$

Finally, we can update the parameters iteratively until convergence, where convergence can be defined as reaching a sufficiently small value of the objective function or when the parameter updates become smaller than a predefined threshold.

##### We derive an iteration formula for regression using Newton’s method instead of gradient descent. showing that the formula reproduces the “normal equations” in regression, this can actually be employed to argue that Newton’s method converges in one iteration in this case:

In this approach, we iteratively update the parameters of the model by approximating the objective function with a quadratic function and then solving for the optimal update.

To derive the iteration formula for regression using Newton's method, we start by defining the objective function as the sum of squared errors between the predicted values and the true values of the response variable:

$f(w) = \sum_{i=1}^{n} (y_i - x_i^T w)^2$

- Initial steps will include: 
computing the gradient vector and the hessian matrix of the objective function: 
$\nabla f(w) = -2X^T(y-Xw)$

$H(w) = 2X^TX$
- Next we update the parameters:

$w^(t+1) = w^(t) - [H(w^(t))]^-1 \nabla f(w^(t))$ for each t iteration

To show that the formula reproduces the normal equations in regression we set the gradient vector to zero and solve:

$\nabla f(w) = -2X^T(y-Xw) = 0$ hence 

$X^TXw = X^Ty$

##### In this case, Newton's method converges in one iteration because the objective function is quadratic and its Hessian matrix is constant. Therefore, the approximation of the objective function at the current parameter values is exact, and the update formula produces the optimal solution in one step.
